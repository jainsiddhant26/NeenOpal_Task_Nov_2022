#### Provide solution and approach for the following:

#### 1. How to identify the Records (Order ID + Product ID combination) present in data1 but missing in data2 (Specify the number of records missing in your answer)

```python
## merging the 2 data sets
```

```python
data_merge_1 = pd.merge(left=data1,right=data2,on=['Order ID','Product ID'],how='left',indicator=True,suffixes=['- data1','- data2'])
```

```python
data_merge_1.head()
```

```python
data_merge_1[data_merge_1['_merge'] == 'left_only']
```

```python
data_merge_1[data_merge_1['_merge'] == 'left_only'].count()
```

##### the number of records missing = 507

#### 2. How to identify the Records (Order ID + Product ID combination) missing in data1 but present in data2 (Specify the number of records missing in your answer)

```python
data_merge_2 = pd.merge(left=data1,right=data2,on=['Order ID','Product ID'],how='right',indicator=True,suffixes=['- data1','- data2'])
```

```python
data_merge_2.head(50)
```

```python
data_merge_2[data_merge_2['_merge'] == 'right_only']
```

```python
data_merge_2[data_merge_2['_merge'] == 'right_only'].count()
```
